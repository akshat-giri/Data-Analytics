### Let define a function which will give us total count of a feature of perticular state in 12 years 

```python
def count(x,y):
    count = year1.loc[x][y]+ year2.loc[x][y]+ year3.loc[x][y]+ year4.loc[x][y]+ year5.loc[x][y]+ year6.loc[x][y]+ year7.loc[x][y]+ year8.loc[x][y]+year9.loc[x][y]+ year10.loc[x][y]+ year11.loc[x][y]+ year12.loc[x][y]
    return count
```

```python
# checking for ANDHRA PRADESH state total Rape cases
count('ANDHRA PRADESH','Rape')
```

### creating a function which create a dataframe with state and feature as an input

```python
def feature_df(x,y):
    feature_df = pd.DataFrame({'year':[2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012]})
    a=year1.loc[x][y]
    b=year2.loc[x][y]
    c=year3.loc[x][y]
    d=year4.loc[x][y]
    e=year5.loc[x][y]
    f=year6.loc[x][y]
    g=year7.loc[x][y]
    h=year8.loc[x][y]
    j=year9.loc[x][y]
    k=year10.loc[x][y]
    l=year11.loc[x][y]
    m=year12.loc[x][y]
    feature = pd.DataFrame({'case_count':[a,b,c,d,e,f,g,h,j,k,l,m]})
    new_df = pd.concat([feature_df,feature], 1)
    return new_df
```

```python
# checking for ANDHRA PRADESH state total Rape cases
feature_df('ANDHRA PRADESH','Rape')
```

### creating a function which will form a dataframe of top 5 state having high count of taken feature

```python
def top_5(y):
    i=0
    z= len(year1.index) - 1
    a = None
    b = None
    c = None
    d = None
    e = None
    cn1 = None
    cn2 = None
    cn3 = None
    cn4 = None
    cn5 = None
    while(i<=z):
        cn= year1.index[i]
        count = year1.loc[cn][y]+ year2.loc[cn][y]+ year3.loc[cn][y]+ year4.loc[cn][y]+ year5.loc[cn][y]+ year6.loc[cn][y]+ year7.loc[cn][y]+ year8.loc[cn][y]+year9.loc[cn][y]+ year10.loc[cn][y]+ year11.loc[cn][y]+ year12.loc[cn][y]
        if(a==None):
            a=count
            cn1 = cn
        elif(b==None):
            b= count
            cn2= cn
        elif(c==None):
            c= count
            cn3= cn
        elif(d==None):
            d= count
            cn4= cn
        elif(e==None):
            e= count
            cn5= cn
        else:
            if(count>= a):
                a= count
                cn1 = cn
            elif(count>=b):
                b= count
                cn2= cn
            elif(count>= c):
                c= count
                cn3= cn
            elif(count>= d):
                d= count
                cn4= cn
            elif(count>= e):
                e= count
                cn5= cn
        i+=1
    Top_5_state = pd.DataFrame({'State':[cn1,cn2,cn3,cn4,cn5],
                            'Count':[a,b,c,d,e]})
    return Top_5_state
```

```python
# let check for Rape cases
top_5('Rape')
```
